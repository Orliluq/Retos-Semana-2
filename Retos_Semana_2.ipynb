## üèÜ **Reto #1: Two Sum  @everyone**

### üìå **Dificultad:** üü¢ F√°cil

## üîπ **Enunciado:**
Dado un array de enteros nums y un entero target, encuentra los √≠ndices de los dos n√∫meros que suman target.

### üîó **URL:** [Two Sum](https://leetcode.com/problems/two-sum/description/)

## üîπ Reglas:

‚úÖ Cada entrada tendr√° exactamente una soluci√≥n.

‚úÖ No puedes usar el mismo n√∫mero dos veces.

‚úÖ Puedes devolver la respuesta en cualquier orden.

```python
# üîπ Ejemplo:
# Entrada: nums = [2,7,11,15], target = 9
# Salida: [0,1]  // Porque nums[0] + nums[1] = 2 + 7 = 9

# Entrada: nums = [3,2,4], target = 6
# Salida: [1,2]  // Porque nums[1] + nums[2] = 2 + 4 = 6
# Entrada: nums = [3,2,4], target = 6
# Salida: [1,2]  // Porque nums[1] + nums[2] = 2 + 4 = 6
```

‚ö†Ô∏è Nota: Lean cuidadosamente las restricciones del problema en LeetCode. üìñ

Este problema nos pide encontrar los √≠ndices de dos n√∫meros en una lista que suman un valor objetivo.

- Creamos un diccionario (num_map) para almacenar los n√∫meros que hemos visto y su √≠ndice.
- Recorremos la lista y, en cada paso, calculamos cu√°nto falta (complement = target - num) para llegar al objetivo.
- Si el n√∫mero que falta ya est√° en el diccionario, significa que encontramos los dos n√∫meros.
- Retornamos los √≠ndices de ambos.

```python
# Ejemplo:
nums = [2, 7, 11, 15]
target = 9
```

- 2 se guarda en el diccionario con √≠ndice 0.
- 7 es el n√∫mero que falta (9 - 2 = 7), lo encontramos en el diccionario.
- Retornamos [0,1], porque nums[0] + nums[1] = 2 + 7 = 9.

```python
def two_sum(nums, target):
    num_map = {}  # Diccionario para almacenar n√∫meros y sus √≠ndices
    for i, num in enumerate(nums):
        complement = target - num  # Calculamos cu√°nto falta para llegar al target
        if complement in num_map:  # Si ya vimos ese n√∫mero antes, devolvemos los √≠ndices
            return [num_map[complement], i]
        num_map[num] = i  # Guardamos el n√∫mero y su √≠ndice en el diccionario

# Ejemplo de uso:
nums = [2, 7, 11, 15]
target = 9
print(two_sum(nums, target))  # Salida: [0, 1]
```

```python
# Ejemplo 2:
print(two_sum([3, 2, 4], 6))  # Salida: [1, 2] (2 + 4 = 6)

# Ejemplo 3:
print(two_sum([1, 5, 3, 7, 8], 10))  # Salida: [1, 3] (5 + 5 = 10)
```

## üèÜ **Reto #9: Palindrome Number**

### üìå **Dificultad:** üü¢ F√°cil

## üîπ **Enunciado:**
Dado un n√∫mero entero x, determina si es un pal√≠ndromo. Un n√∫mero es un pal√≠ndromo si se lee igual de izquierda a derecha y de derecha a izquierda.

### üîó **URL:** [Palindrome Number](https://leetcode.com/problems/palindrome-number/description/)

## üîπ **Reglas:**
‚úÖ No puedes convertir x a una cadena (string).

‚úÖ Debes resolverlo utilizando operaciones matem√°ticas.

```python
#üîπ Ejemplo:
# Entrada: x = 121
# Salida: true
# Explicaci√≥n: 121 le√≠do al rev√©s sigue siendo 121.

# Entrada: x = -121
# Salida: false
# Explicaci√≥n: -121 le√≠do al rev√©s es 121-, no es igual al original.

# Entrada: x = 10
# Salida: false
# Explicaci√≥n: 10 le√≠do al rev√©s es 01, no es un pal√≠ndromo.
```

‚ö†Ô∏è Nota: Lean cuidadosamente las restricciones del problema en LeetCode. üìñ

Este problema nos pide verificar si un n√∫mero se lee igual de izquierda a derecha y de derecha a izquierda sin convertirlo en string.

- Si el n√∫mero es negativo, devolvemos False porque los negativos no pueden ser pal√≠ndromos (-121 le√≠do al rev√©s es 121-).
- Invertimos el n√∫mero poco a poco extrayendo el √∫ltimo d√≠gito y construyendo el n√∫mero al rev√©s.
- Comparamos el n√∫mero original con el invertido. Si son iguales, es un pal√≠ndromo.

```python
# Ejemplo:
x = 121
```

- Invertimos 121 ‚Üí 121, es igual al original ‚Üí True.
- -121 al rev√©s es 121-, diferente ‚Üí False.
- 10 al rev√©s es 01, diferente ‚Üí False.

```python
def is_palindrome(x):
    if x < 0:
        return False  # Los negativos no pueden ser pal√≠ndromos
    reversed_num = 0
    original_num = x
    while x > 0:
        reversed_num = reversed_num * 10 + x % 10  # Extraemos el √∫ltimo d√≠gito y lo agregamos al n√∫mero invertido
        x //= 10  # Eliminamos el √∫ltimo d√≠gito del n√∫mero original
    return original_num == reversed_num  # Comparamos el original con el invertido

# Ejemplo de uso:
x = 121
print(is_palindrome(x))  # Salida: True
```

```python
# Ejemplo 2:
print(is_palindrome(-121))  # Salida: False (-121 al rev√©s es 121-)

# Ejemplo 3:
print(is_palindrome(10))  # Salida: False (10 al rev√©s es 01)
```

## üèÜ **Reto #167: Two Sum II - Input Array is Sorted**

### üìå **Dificultad:** üü° Media

## üîπ **Enunciado:**
Dado un array ordenado de enteros numbers y un entero target, encuentra los √≠ndices de los dos n√∫meros que suman target.

## üìå **Diferencia con el problema original (Two Sum):**
‚úÖ El array est√° ordenado en orden ascendente.

‚úÖ Debes usar una soluci√≥n m√°s eficiente que la de fuerza bruta.

‚úÖ Los √≠ndices empiezan desde 1 (no desde 0).

### üîó **URL:** [Two Sum II - Input Array Is Sorted](https://leetcode.com/problems/two-sum-ii-input-array-is-sorted/description/)

## üîπ **Reglas:**
‚úÖ No puedes modificar el array original.

‚úÖ Debes devolver los √≠ndices en base 1 (es decir, el primer elemento tiene √≠ndice 1, no 0).

```python
# üîπ Ejemplo:
# Entrada: numbers = [2,7,11,15], target = 9
# Salida: [1,2]  // Porque numbers[1] + numbers[2] = 2 + 7 = 9

# Entrada: numbers = [-1,0], target = -1
# Salida: [1,2]  // Porque numbers[1] + numbers[2] = -1 + 0 = -1
```

## üîπ **Pistas:**
### üí° Dos son mejores que unoüëÄ

### ‚ö†Ô∏è Nota: Aseg√∫rate de leer las restricciones del problema en LeetCode. üìñ

Este problema es una versi√≥n optimizada del primero, pero el array ya est√° ordenado y debemos encontrar los n√∫meros de forma eficiente.

- Usamos dos punteros: uno al inicio (left) y otro al final (right) del array.
Sumamos los n√∫meros en esas posiciones:
- Si la suma es menor que el objetivo, movemos left hacia la derecha para aumentar la suma.
- Si la suma es mayor, movemos right hacia la izquierda para disminuir la suma.
- Si encontramos la suma exacta, devolvemos los √≠ndices (+1 porque en este problema los √≠ndices empiezan en 1, no en 0).

```python
# Ejemplo:
numbers = [2, 7, 11, 15]
target = 9
```

- numbers[0] + numbers[1] = 2 + 7 = 9 üéØ
- Retornamos [1,2] (porque los √≠ndices en este problema empiezan en 1 y no en 0).

```python
def two_sum_sorted(numbers, target):
    left, right = 0, len(numbers) - 1  # Punteros al inicio y final del array
    while left < right:
        total = numbers[left] + numbers[right]
        if total == target:
            return [left + 1, right + 1]  # +1 porque los √≠ndices comienzan en 1
        elif total < target:
            left += 1  # Si la suma es menor, movemos el puntero izquierdo hacia la derecha
        else:
            right -= 1  # Si la suma es mayor, movemos el puntero derecho hacia la izquierda

# Ejemplo de uso:
numbers = [2, 7, 11, 15]
target = 9
print(two_sum_sorted(numbers, target))  # Salida: [1, 2]
```

```python
# Ejemplo 2:
print(two_sum_sorted([-1, 0], -1))  # Salida: [1, 2] (-1 + 0 = -1)

# Ejemplo 3:
print(two_sum_sorted([1, 3, 4, 5, 7, 10, 11], 9))  # Salida: [3, 5] (4 + 5 = 9)
```

```python
import gradio as gr

def two_sum(nums, target):
    num_map = {}
    nums = list(map(int, nums.split(",")))  # Convertir texto a lista de enteros
    for i, num in enumerate(nums):
        complement = target - num
        if complement in num_map:
            return f"√çndices encontrados: {num_map[complement]}, {i}"
        num_map[num] = i
    return "No se encontraron √≠ndices"

# Crear la interfaz con Gradio
demo = gr.Interface(
    fn=two_sum,
    inputs=["text", "number"],
    outputs="text",
    title="Two Sum Finder",
    description="Ingresa una lista de n√∫meros separados por comas y un objetivo."
)

# Mostrar interfaz
demo.launch(share=True)  # share=True para obtener un link p√∫blico
```

```python
import gradio as gr

def two_sum_sorted(numbers, target):
    numbers = list(map(int, numbers.split(",")))  # Convertir texto a lista de enteros
    left, right = 0, len(numbers) - 1
    while left < right:
        sum_ = numbers[left] + numbers[right]
        if sum_ == target:
            return f"√çndices encontrados: {left+1}, {right+1}"  # Base 1
        elif sum_ < target:
            left += 1
        else:
            right -= 1
    return "No se encontraron √≠ndices"

# Crear la interfaz en Gradio
demo = gr.Interface(
    fn=two_sum_sorted,
    inputs=["text", "number"],
    outputs="text",
    title="Two Sum II - Sorted Array",
    description="Ingresa una lista de n√∫meros ordenados y un objetivo."
)

demo.launch(share=True)
```
